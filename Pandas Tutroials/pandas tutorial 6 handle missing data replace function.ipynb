```python
#using replace function
import pandas as pd
import numpy as np
df = pd.read_csv("weather_data_tutorial_6.csv")
df
```

```python
#replece values based on specific columns
new_df = df.replace({
    'temperature' : -99999,
    'windspeed' : -88888,
    'event' : '0'
}, np.NaN)
new_df
```

```python
new_df = df.replace({
    -99999: np.NaN, 
    'No Event' : 'Sunny'
})
new_df
```

```python
#dealing with units of measure based on columns by using a dictionary
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
}, '', regex= True)
new_df
```

```python
#replacing a list of values with another list of values
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
new_df = df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
new_df
```
