```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "baltimore"],
    "temperature": [21, 14, 35, 32]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago", "new york", "san francisco"],
    "humidity": [65, 68, 71]
})

df2
```

```python
#joining df1 and df2 using merge
df3 = pd.merge(df1, df2, on="city")
df3

#intersection or inner join occures,similar components are taken,how = "inner"
```

```python
# for a union ,use how = "outer"
df4 = pd.merge(df1, df2, on="city", how= 'outer')
df4
```

```python
#how="left",takes elements of the left set
# order decided by how you call the dfs,IOC df1,df2,
# df1 = left
df5 = pd.merge(df1, df2, on="city", how= 'left')
df5
```

```python
#how = "right"
df5 = pd.merge(df1, df2, on="city", how= 'right')
df5
```

```python
# to know where data came from,indicator = True
df4 = pd.merge(df1, df2, on="city", how= 'outer', indicator= True)
df4
```

```python
#suffixes
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
pd.merge(df1, df2, on = "city", suffixes= ('_left', '_right'))
```
