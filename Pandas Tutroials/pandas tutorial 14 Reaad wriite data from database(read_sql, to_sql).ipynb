# writing data to df from sql tables


```python
import pandas as pd
import sqlalchemy
import pymysql
```

```python
# create engine
engine = sqlalchemy.create_engine('mysql+pymysql://root:2401@localhost/kimanidb')
```

```python
# reading customers table from kimanidb
df = pd.read_sql_table("customers", engine)
df
```

```python
# reading specific columns(name and phone number)
df = pd.read_sql_table("customers", engine, columns= ["name", "phone_number"])
df
```

```python
# join customers and orders tables and read into a data frame using pd.read_sql_query()
query='''
SELECT customers.name, customers.phone_number, orders.name, orders.amount
FROM customers INNER JOIN orders
ON customers.id=orders.id
'''
df = pd.read_sql_query(query, engine)
df
```

# writing data from df to sqltable

```python
# read csv and load data to data frame
df = pd.read_csv("customers_tutorial_14.csv")
df
```

```python
# write to sql table
# first rename columns in df
df.rename(columns = {
    'Customer Name': 'name',
    'Customer Phone': 'phone_number'
}, inplace = True)
df
```

```python
# write to sql table
df.to_sql(
    name = 'customers',
    con =  engine,
    index = False,
    if_exists = 'append'
)
```
