```python
#pivot allows you to reshape or transform yout dataframe
import pandas as pd
df = pd.read_csv("weather_tutorial_10.csv")
df
```

```python
# transforming df
df.pivot(index = "humidity", columns = "city")
```

```python
# pivot table summarizes and aggregates data inside a dataframe
df = pd.read_csv("weather2_tutorial_10.csv")
df
```

```python
# average temperature in day
df.pivot_table(index = "city", columns = "date", margins = True)
```

```python
df = pd.read_csv("weather3_tutorial_10.csv")
df
```

```python
# grouper() aggregates based on date frequency
df['date'] = pd.to_datetime(df['date'])  #convert date from string to date
df

```

```python
df.pivot_table(index = pd.Grouper(freq = "M", key= "date"), columns="city")
```
