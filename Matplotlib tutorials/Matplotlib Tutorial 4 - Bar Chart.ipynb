# Matplotlib Tutorial 4 - Bar Chart


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company= ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue= [90, 136, 89, 27]
```

```python
xpos = np.arange(len(company))
# convert company from string to int
xpos
```

```python
# replacing numbers with company names
# plt.xticks()
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos, revenue)
```

# Multipple bars

```python
# plotting two diifferent measures
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos, revenue, label="Revenue")
plt.legend()
```

```python
# adding profit list
profit=[40,2,34,12]
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos, revenue, label="Revenue")
plt.bar(xpos, profit, label= "Profit")
plt.legend()
# plots on the same bar
```

```python
# plotting side by side
# xpos- or + and width = 0.4
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2, revenue, width= 0.4, label="Revenue")
plt.bar(xpos+0.2, profit, width= 0.4, label= "Profit")
plt.legend()
```

# Horizontal bar chart

```python
# use function barh(), use yticks
plt.yticks(xpos, company)

plt.title("US Tech Stocks")

plt.barh(xpos-0.2, revenue, label="Revenue")
plt.barh(xpos+0.2, profit, label= "Profit")
plt.legend()
```
