# Matplotlib Tutorial 6 - Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
```

```python
plt.axis("equal")  #circular impression
plt.pie(exp_vals, labels= exp_labels, radius= 1.5, autopct= '%0.1f%%', shadow= True)#change size radius = ---,for percentage,autopct= '%0.2f%%',2dp
plt.show()
```

```python
# explode
# highlite a pie
plt.axis("equal")  
plt.pie(exp_vals, labels= exp_labels, radius= 1.5, autopct= '%0.1f%%', shadow= True, explode= [0,0.1,0.5,0,0])
plt.show()
```

```python
# change start angle
plt.axis("equal")  
plt.pie(exp_vals, labels= exp_labels, radius= 1.5, autopct= '%0.1f%%', shadow= True, explode= [0,0.1,0.5,0,0], startangle= 180)
plt.show()
```

# saving file

```python
# use plt.savefig()
fig1 = plt.gcf()
plt.axis("equal")  
plt.pie(exp_vals, labels= exp_labels, radius= 1.5, autopct= '%0.1f%%', shadow= True, explode= [0,0.1,0.5,0,0], startangle= 180)
plt.show()
fig1.savefig("piechart.pdf", bbox_inches= "tight", pad_inches= 2, transparent= True)
# bbox_inches = "tight", ensures fits in png file
# transparent= True removes image background
```
