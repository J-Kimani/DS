# Tutorial 3:Axes labels, Legend, Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
# ploting, plt.plot(x-axis, y-axis)
# plotting three variables on same plot
plt.xlabel("Days")
plt.xlabel("Temperature")
plt.title("Weather")
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

# assigning lables to lines

```python
plt.xlabel("Days")
plt.xlabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label= "Max")
plt.plot(days, min_t, label= "Min")
plt.plot(days, avg_t, label= "Avg")
```

```python
# show legend
plt.xlabel("Days")
plt.xlabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label= "Max")
plt.plot(days, min_t, label= "Min")
plt.plot(days, avg_t, label= "Avg")

plt.legend()
```

```python
# changing location of legend
plt.xlabel("Days")
plt.xlabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label= "Max")
plt.plot(days, min_t, label= "Min")
plt.plot(days, avg_t, label= "Avg")

plt.legend(loc= "upper right")
# default is plt.legend(loc= "best")
```

```python
days = [1,2,3,4,5,6,7]
max_t = [32,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]


plt.xlabel("Days")
plt.xlabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label= "Max")
plt.plot(days, min_t, label= "Min")
plt.plot(days, avg_t, label= "Avg")

plt.legend()
```

```python
# adding a shadow property, shadow = "True"
# and other properties
plt.xlabel("Days")
plt.xlabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label= "Max")
plt.plot(days, min_t, label= "Min")
plt.plot(days, avg_t, label= "Avg")

plt.legend(loc= "best", shadow= True, fontsize= "small")
```
