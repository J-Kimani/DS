# Matplotlib tutorial 1


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# x= number of weekdays
# y=temperature value
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
# assigning x and y label
plt.xlabel('Day')
plt.ylabel('Tempertaure')
plt.title('Weather')
plt.plot(x,y, color= 'green', linewidth= 5, linestyle= 'dotted')
```
