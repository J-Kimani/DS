# Matplotlib Tutorial 5 - Histograms

## Histograms are barcharts where x axis contains frequency of a variable

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
# drawing the histogram
# bins function is assigning a range
# make bars distinguishable using rwidth=

plt.hist(blood_sugar, bins=3, rwidth= 0.95)
```

# Bins Function

```python
# bin parameter can be changed by supplying a list
plt.hist(blood_sugar, bins= [80, 100, 125, 150], rwidth= 0.95)
```

```python
# api for hist() function
# color= ,histtype= step,
plt.hist(blood_sugar, bins=3, rwidth= 0.95, color= 'r', histtype= 'step')
```

```python
plt.hist(blood_sugar, bins=3, rwidth= 0.95,color= 'indigo')
```

# working with multiple samples


```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('sugar range')
plt.ylabel('Total number of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=3, rwidth= 0.95,color= ['indigo', 'pink'],
        label= ['men', 'women'])
plt.legend()
```

# oreintation property

```python
plt.xlabel('sugar range')
plt.ylabel('Total number of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=3, rwidth= 0.95,color= ['indigo', 'pink'],
        label= ['men', 'women'], orientation= 'horizontal')
plt.legend()
```
