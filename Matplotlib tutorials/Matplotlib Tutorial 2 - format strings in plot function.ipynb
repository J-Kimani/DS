# Tutorial 2:format strings in plot function

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# x= number of weekdays
# y=temperature value
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
# PLOTTING
# format strings,g+
plt.plot(x,y, 'rD--')
```

```python
# using key word arguments
# markersize = 55,controls marker size
plt.plot(x, y, color= 'red', marker= 'D', linestyle= '', markersize= 20)
```

```python
plt.plot(x, y, color= 'red', marker= '+', linestyle= '', markersize= 20)
```

```python
# alpha = values between 0-1,
# controls transarency of curve

plt.plot(x, y, color= 'red', alpha= 0.2)
```
