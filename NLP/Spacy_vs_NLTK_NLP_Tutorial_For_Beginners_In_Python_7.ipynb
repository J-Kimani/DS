sentence and word tockenization in spacy

```python
import spacy
```

```python
# sentence tockenization
nlp = spacy.load("en_core_web_sm")

doc = nlp("Dr. Strange loves pav bhaji of mumbai. Hulk loves chaat of delhi")

for sentence in doc.sents:
  print(sentence)
```

```python
for sentence in doc.sents:
  for word in sentence:
    print(word)
```

sentence and word tockenization in nltk

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize

sent_tokenize("Dr. Strange loves pav bhaji of mumbai. Hulk loves chaat of delhi")
```

```python
from nltk.tokenize import word_tokenize

word_tokenize("Dr. Strange loves pav bhaji of mumbai. Hulk loves chaat of delhi")
```
