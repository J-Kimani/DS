# 3. Largest Producers In Last 5 Years

first see how initial top 5 compare with latest top 5

```python
import numpy as np
```

```python
def top_rank(r, n= 5):
    r = r.sort_values(by= "Total Emissions", ascending= False).head(n)["Country"]
    r.index = np.arange(len(r)) + 1
    r.index = "Rank #" + r.index.astype(str)
    r = r.transpose()
    return r
```

```python
top_rank(df2.query('Year == "1990"'))
```

```python
# top 5 largest producers 2014-2018
df2[df2.Country != "World"].groupby("Year").apply(top_rank).sort_index(ascending=False)[:5]
```

# *Observations*

1. Top 5 countries remains quite consistent except for Brazil which replaces Indonesia in the top 5

2. Russia drops to second pushing the United States to third with China taking first place

# 5. Largest contributors in 90s and 2000s


```python
len(df2.Year.unique())
```

```python
#  see how thing were in the 90s
df2[df2.Country != "World"].groupby("Year").apply(top_rank).sort_index()[:10]
```

# *Observations*
1. China's production increase to a point it is the highest producer of methane between 1996 - 1999.

2. India maintains a constant rate of production maintaining its position as 4th throughout the 90's.

3. During the first 6 years of the decade, Russia is the leading producer until it is overtaken by China in 1996 falling to 3rd between 1996 and 1997 before displacing the United States in 1998.

4. The United States holds its position for the first 8 years of the decade in 2nd place before its displacement by Russia.

```python
# state of affairs in the 2000s
df2[df2.Country != "World"].groupby("Year").apply(top_rank).sort_index()[10:]
```

# *Observations*

1. Russia and China keep changing positions between 1 and 2 until 2008 when China settles at the top and Russia assumes second place.

2. The United States and India maintain their positions at 3rd and 4th throughout the 2000s.

3. Brazil establishes its position at 5th by displacing Indonesia fully after 2006 probably due to increased industrialization in the country.

# 5. Rankings per Sector

Main interest is the year 2018

```python
df_2018 = df2.query('Year == "2018"')
df_2018.head()
```

```python
import plotly.express as px
from plotly.subplots import make_subplots 
from plotly.graph_objs import graph_objs as go
```

```python
rows = 4
cols = 2

sectors = ["Total Emissions", "Agriculture", "Energy", "Waste", "Land-Use Change and Forestry", "Fugitive Emissions", "Other Fuel Combustion", "Industrial Processes"]

subplot_titles = [l for l in sectors]

specs = [[{'type': 'bar'}] * cols] * rows

fig = make_subplots(
    rows= rows,
    cols= cols,
    subplot_titles= subplot_titles,
    specs= specs
)

for i, s in enumerate(sectors):
    row = i // cols + 1 
    col = i % cols + 1

    d = df_2018.sort_values(by= s, ascending= False)[:5]

    fig.add_trace(
        go.Bar(
            x = d["Country"],
            y = d[s],
            showlegend= False,
            legendgroup= "Countries",
            textposition= "outside",
            texttemplate= "%{y}",
            name= s
        ),
        row= row,
        col= col
    )

for i in range(1, rows * cols + 1):
    fig['layout']['yaxis{}'.format(i)]['title'] = "Emissions (MTCO2e)"


fig.update_layout(title= "Top five Methane Producers By Sector", title_x= 0.5, height= 1500)
fig.show()
```
