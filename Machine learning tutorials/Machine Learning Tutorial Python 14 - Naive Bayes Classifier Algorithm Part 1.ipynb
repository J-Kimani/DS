## Naive Bayes Tutorial Part 1: Predicting survival from titanic crash

```python
import pandas as pd
df = pd.read_csv("titanic_ML14.csv")
df.head()
```

```python
# data exploration
# dropping non_essential columns
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis= 'columns', inplace= True)
df.head()
```

```python
# target varible is survived
# separate
#dependent variable = inputs
target = df.Survived
inputs = df.drop('Survived', axis= 'columns')
```

```python
# sex column is text
# convert to numbers
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
# append to input df
inputs = pd.concat([inputs, dummies], axis= 'columns')
inputs.head()
```

```python
# drop sex column
inputs.drop('Sex', axis= 'columns', inplace= True)
inputs.head()
```

```python
# find out if there are na's
inputs.columns[inputs.isna().any()]
```

```python
# investigate where the missing values are in the first 20 elements
inputs.Age[:20]
```

```python
# handling Nan values
# use mean
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(20)
```

```python
# spliting
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size= 0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(inputs)
```

```python
x_train
```

```python
# using gaussian naivebayes to creat model
# useful for normal distribution
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
# training the model
model.fit(x_train, y_train)
```

```python
# measure accuracy
model.score(x_test, y_test)
```
