```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
from pandas import Series, DataFrame
# create dataframe for visualization

iris_data = DataFrame(X, columns= ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"])

iris_target = DataFrame(y, columns= ["Species"])
```

```python
iris_data.head()
```

```python
iris_target.sample(10)
```

```python
# at the mometn we have 0, 1 and 2 for species, change to name


def flower(num):
    if num == 0:
        return "Setosa"
    elif num == 1:
        return "Versicolour"
    else:
        return "Virginica"
```

```python
# label flowers by combining dataframes

iris_target["Species"] = iris_target["Species"].apply(flower)
```

```python
iris_target.sample(5)
```

```python
# combine dataframes

irisdf = pd.concat([iris_data, iris_target], axis= "columns")
irisdf.head()
```

# train model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

model = LogisticRegression()

model
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state= 3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model.fit(X_train, y_train)
```

```python
# test accuracy
model.score(X_test, y_test)
```

```python
y_test
```

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
# print a confusion matrix
from sklearn import metrics
from sklearn.metrics import confusion_matrix
print(metrics.accuracy_score(y_test, y_predicted))

cm = confusion_matrix(y_test, y_predicted)
cm

```

```python
# confusion_matrix visualization
plt.figure(figsize= (10,7))
sns.heatmap(cm, annot= True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
