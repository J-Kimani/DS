```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income_ML2.csv")
df.sample(5)
```

#  plots

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year, df["per capita income (US$)"], color= "red", marker= "*")
```

```python
plt.plot(df.year, df["per capita income (US$)"], color= "red")
```

```python
plt.step(df.year, df["per capita income (US$)"], color= "red", marker= "*")
```

# linear regression model


```python
# create linear regression model
reg = linear_model.LinearRegression()
reg.fit(df[["year"]], df["per capita income (US$)"])
```

```python
# task : predict income in 2022
# np.asscalar(reg.predict([[2022]]))
# ans = reg.predict([[2022]]).item()
# type(ans)

reg.predict([[2022]]).item()
```

```python
# coefficients
print("coefficent: ", reg.coef_.item())

# intercept
print("intercept: ", reg.intercept_)
```

```python
# draw linear regression line
%matplotlib inline
plt.xlabel("year", fontsize= 15)
plt.ylabel("per capita income (US$)", fontsize= 15)
# plt.scatter(df.year, df["per capita income (US$)"], color= "red", marker= "*")

# adding linear reg line
plt.plot(df.year, reg.predict(df[["per capita income (US$)"]]), color= "green")
```
