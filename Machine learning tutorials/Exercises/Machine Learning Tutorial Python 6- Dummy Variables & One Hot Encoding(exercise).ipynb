First plot data points on a scatter plot chart to see if linear regression model can be applied. If yes, then build a model that can answer following questions,

1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

use carprices_exercise_ML6.csv

```python
import pandas as pd
df = pd.read_csv("carprices_exercise_ML6.csv")
df
```

create dummy variables column

```python
dummy = pd.get_dummies(df["Car Model"])
dummy
```

```python
# combining dummy with oriiginal df
merged = pd.concat([df, dummy], axis= "columns")
merged
```

```python
# car model column dropped
final = merged.drop(["Car Model"], axis= "columns")
final
```

```python
# createlinear model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# definr x and y for training
X = final.drop("Sell Price($)", axis= "columns")
X
```

```python
y= final["Sell Price($)"]
y
```

```python
# training
model.fit(X, y)
```

1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

```python
# prediction
# mercedez benz that is 4 yr old with mileage 45000
model.predict([[45000, 4, 0, 0, 1]])
```

```python
# BMW X5 that is 7 yr old with mileage 86000
model.predict([[86000, 7, 0, 1, 0]])
```

```python
model.score(X, y)
```

# Using sklearn one hot encoder

```python
df
```

```python
# label encoding on car model column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# new df
dfle = df
dfle["Car Model"] = le.fit_transform(dfle["Car Model"])
dfle
```

```python
# create X and y
X = df[["Car Model", "Mileage", "Age(yrs)"]].values
X
```

```python
y = dfle["Sell Price($)"]
y
```

```python
# use one hot encoder to create dummmies for each vehicle
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
one = ColumnTransformer([("Car Model", OneHotEncoder(), [0])], remainder= "passthrough")
one
```

```python
X = one.fit_transform(X)
X
```

```python
# dropping using dummy variable trap
X = X[:, 1:]
X
# first take all rows, then all colimns from index one onwards
```

```python
# training
model.fit(X, y)
```
